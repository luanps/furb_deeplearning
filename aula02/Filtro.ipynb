```python
import sys
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = 'image.jpg'
image = cv2.imread(img_path)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
some_filter = np.array([[ -1, -2, -1],
                       [ 0, 0, 0],
                       [ 1, 2, 1]])
```

```python
filtered_img = cv2.filter2D(gray, -1, some_filter)
```

```python
plt.imshow(filtered_img, cmap='gray')
```
